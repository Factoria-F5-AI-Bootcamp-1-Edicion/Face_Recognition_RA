```python
import cv2
import dlib
import face_recognition
import numpy as np

```

```python
# Cargamos imagen
image = cv2.imread("Rebeca.JPG")
```

```python
face_loc = face_recognition.face_locations(image)[0]
```

```python
print("face loc:", face_loc)
```

```python
face_image_encodings = face_recognition.face_encodings(image, known_face_locations=[face_loc])[0]
```

```python
print(" face image encodings: ", face_image_encodings)
```

Reconocimiento facial mediante video

```python
cap = cv2.VideoCapture(1)
```

```python
while True:
    ret, frame = cap.read()
    if ret == False : break
    frame = cv2.flip(frame, 1)
  
    face_locations = face_recognition.face_locations(frame)
    if face_locations !=[]:
        for face_locations in face_locations:
            face_frame_encodings = face_recognition.face_encodings(frame, known_face_locations=[face_locations])[0]
            result = face_recognition.compare_faces([face_image_encodings], face_frame_encodings)
            print("Result: ", result)
            
            if result[0] == True:
                text = "Rebe"
                color = (125,220,0)
            else:
                text = "Desconocido"
                color = (50,50, 255)
                
            cv2.rectangle(frame, (face_locations[3], face_locations[2]),(face_locations[1], face_locations[2] +30), color, -1)
            cv2.rectangle(frame, (face_locations[3], face_locations[0]),(face_locations[1], face_locations[2]), color, 2)
            cv2.putText(frame, text, (face_locations[3], face_locations[2] + 20), 2, 0.7 ,(255,255,255), 1)

    cv2.imshow("Frame" , frame)
    k = cv2.waitKey(1)
    if k == 27 & 0xFF:
        break
```
